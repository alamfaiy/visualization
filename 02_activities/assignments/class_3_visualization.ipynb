```python
# load libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
import PIL
import requests
```

```python
# sample data
np.random.seed(613)
x = np.arange(50)
y1 = np.random.randint(0,100,50)
y2 = np.random.randint(0,100,50)

fig,ax = plt.subplots(figsize = (5,3))

# Plotting two of these y values on the same axis 
ax.plot(x,y1) 
ax.plot(x,y2)

#fig.show() # can use for reproducibility purposes because some IDEs don't show the graph without this
```

```python
# Define plot again
fig,ax = plt.subplots(figsize = (5,3))

# Plotting two of these y values on the same axis 
ax.plot(x,y1, label = "Person 1") # these labels are necessary for the legend later
ax.plot(x,y2, label = "Person 2")

# Lets add a legend now
ax.legend(loc='lower right')

```

```python
# Define plot again
fig,ax = plt.subplots(figsize = (5,3))

# Plotting two of these y values on the same axis 
ax.plot(x,y1, label = "Person 1") 
ax.plot(x,y2, label = "Person 2")

# Legend (we will modify this now)
ax.legend(loc='lower right',
          frameon = True, #add a frame around the legend)
          fontsize = 12, #change font size,
          ncol = 2, #specify number of columns
          shadow = True)
```

```python
# Now we can move the legend outside of the plot area
# Define plot again
fig,ax = plt.subplots(figsize = (5,3))

# Plotting two of these y values on the same axis 
ax.plot(x,y1, label = "Person 1") 
ax.plot(x,y2, label = "Person 2")

# Legend (we will modify this again to put outside)
ax.legend(loc='upper left',
          bbox_to_anchor = (0.5,0.5)) 
# loc picks a corner of the legend as an anchor point
# bbox_to_anchor picks a location for that anchor 
```

```python
# Annotating with the text() function

# Define plot again
fig,ax = plt.subplots(figsize = (5,3))

# Making a scatter plot
ax.scatter(x,y1, label = "Person 1") 
ax.scatter(x,y2, label = "Person 2")

# Legend
ax.legend(loc='lower right')

# Modifying text annotations
ax.text(10,95,"This value is important!",
        ha = 'center',
        color = 'red',
        size = 20)
```

```python
# Positioning text annotations using transforms

fig, ax = plt.subplots()
ax.axis([0, 10, 0, 10]) # the x and y range is in square brackets

ax.text(1,5, ". Data:(1,5)", transform = ax.transData)
ax.text(0.5, 0.1, ". Axes:(0.5, 0.1)", transform = ax.transAxes)
ax.text(0.2, 0.2, ". Figure:(0.2, 0.2)", transform = fig.transFigure)
```

```python
fig, ax = plt.subplots(figsize=(5,3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "person 2")
ax.legend(loc="lower right")

ax.annotate("This is important!", 
            xy=(10,95), 
            xytext=(20,94),
            arrowprops = dict(arrowstyle = "wedge",
                              color = "hotpink"))
```

```python
# Removing tick marks and labels 

fig, ax = plt.subplots(figsize=(5,3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "person 2")
ax.legend(loc="lower right")

ax.yaxis.set_major_locator(plt.NullLocator())
ax.xaxis.set_major_formatter(plt.NullFormatter())

# Ooo all those marks and labels really do be removed
```

```python
fig, ax = plt.subplots(figsize=(5,3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc = "lower right")

# This limits the tick marks we can see
#ax.xaxis.set_major_locator(plt.MaxNLocator(3))
ax.xaxis.set_major_locator(plt.MaxNLocator(5))
```

```python
fig, ax = plt.subplots(figsize=(5,3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc = "lower right")

# This indicates we want tick marks at intervals of 5
ax.xaxis.set_major_locator(plt.MultipleLocator(5))
```

```python
# Continue to modify tick marks
# Let's rotate them

fig, ax = plt.subplots(figsize=(5,3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc = "lower right")

# This will rotate them
# plt.xticks(rotation=45, ha="right")
plt.xticks(rotation=45, ha="center") # this looks better
```
